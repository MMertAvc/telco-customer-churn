# Telco Customer Churn - AutoML

# Introduction

This notebook showcases an AutoML pipeline using PyCaret for predicting customer churn from the Telco dataset.

We aim to:
- Understand key customer attributes
- Automatically train multiple models
- Predict the `Churn` status for new customers

The target variable is `Churn` (Yes/No).

```python
import pandas as pd
from pycaret.classification import *
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.columns = train.columns.str.strip()  # boşluk varsa temizle
train.columns
```

```python
s = setup(data=train, target='Churn', session_id=42, verbose=False)
```

```python
best_model = compare_models()
```

```python
preds = predict_model(best_model, data=test)
```

```python
submission = pd.DataFrame({'Churn_Predicted': preds['prediction_label']})
```

```python
submission.to_csv('submission.csv', index=False)
```

```python
save_model(best_model, 'best_automl_model') 
```

# Conclusion

Through PyCaret's AutoML capabilities, this notebook efficiently predicts customer churn for telecom services.

The workflow includes:
- Data preparation
- Model comparison
- Submission generation

This end-to-end process is integrated into a deployable Streamlit app for real-time usage.
